```python
from datasets import load_dataset

dataset = load_dataset("Cynaptics/Test2Sql")
```

```python
dataset
```

```python
def num_of_tables(split):
  table_count_list=[]
  for schema in dataset[split]['schema']:
    table_count_list.append(schema.count('CREATE TABLE'))
  return table_count_list
```

```python
table_count_list = num_of_tables('small_train')
table_count_list
```

```python
from collections import Counter
count_frequency = Counter(table_count_list)
count_frequency
```

```python
max_num_tables = max(table_count_list)
max_num_tables
```

```python
max_table_indexes = [i for i, x in enumerate(table_count_list) if x == max_num_tables]
max_table_indexes
```
